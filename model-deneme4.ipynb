```python
non_numeric_amounts = df[df['Amount'].apply(lambda x: pd.to_numeric(x, errors='coerce')).isnull()]
df_cleaned = df.drop(non_numeric_amounts.index)
df_cleaned['Amount'] = pd.to_numeric(df_cleaned['Amount'], errors='coerce')
df_cleaned_info = df_cleaned.info()
```

```python
df_cleaned = df_cleaned.dropna(subset=['Pan', 'CardBrandName', 'TransactionRequestDate'])
df_cleaned.isnull().sum()
```

```python
import numpy as np
from datetime import datetime, timedelta

def create_synthetic_data(valid_data, num_synthetic):
    synthetic_list = []
    for i in range(num_synthetic):
        sample = valid_data.sample(n=1, random_state=i)
        original_amount = sample['Amount'].values[0]
        sapma_orani = np.random.uniform(0.1, 0.3)
        amount_variation = original_amount * sapma_orani
        new_amount = original_amount + np.random.uniform(-amount_variation, amount_variation)
        random_minutes = np.random.uniform(1, 4 * 24 * 60)
        new_date = sample['CreateDate'].values[0] + pd.Timedelta(minutes=int(random_minutes))
        request_date = new_date + pd.Timedelta(seconds=1)
        response_date = request_date + pd.Timedelta(seconds=1)
        pan_prefix = sample['Pan'].values[0][0:6] if pd.notnull(sample['Pan'].values[0]) else '000000'
        random_pan = pan_prefix + '******' + str(np.random.randint(1000, 9999)).zfill(4)
        card_brand = sample['CardBrandName'].values[0] if pd.notnull(sample['CardBrandName'].values[0]) else 'Unknown'
        card_bank = sample['CardBankName'].values[0] if pd.notnull(sample['CardBankName'].values[0]) else 'Unknown'
        card_family = sample['CardFamily'].values[0] if pd.notnull(sample['CardFamily'].values[0]) else 'NoFamily'
        card_type = sample['CardType'].values[0] if pd.notnull(sample['CardType'].values[0]) else 'CREDIT'
        synthetic_list.append({
            'Pan': random_pan,
            'Amount': new_amount,
            'CreateDate': new_date,
            'TransactionRequestDate': request_date,
            'TransactionResponseDate': response_date,
            'TransactionStatus': sample['TransactionStatus'].values[0],
            'TransactionType': sample['TransactionType'].values[0],
            'CardBrandName': card_brand,
            'CardBankName': card_bank,
            'CardFamily': card_family,
            'CardType': card_type,
            'Fraud': 0
        })
    return pd.DataFrame(synthetic_list)

valid_data = df_cleaned[df_cleaned['Fraud'] == 0]
synthetic_valid_df = create_synthetic_data(valid_data, 15000)
df_augmented = pd.concat([df_cleaned, synthetic_valid_df], ignore_index=True)
```

```python
def create_fraudulent_groups(valid_data, num_fraud_groups, num_transactions_per_group_range):
    fraud_list = []
    for i in range(num_fraud_groups):
        sample = valid_data.sample(n=1, random_state=i)
        pan = sample['Pan'].values[0]
        card_brand = sample['CardBrandName'].values[0]
        card_bank = sample['CardBankName'].values[0]
        card_family = sample['CardFamily'].values[0]
        card_type = sample['CardType'].values[0]
        
        # Başlangıç miktarı ve tarih belirleme
        start_amount = np.random.uniform(100, 5000)  # Sabit kalacak miktar
        start_date = sample['CreateDate'].values[0]
        num_transactions_per_group = np.random.randint(*num_transactions_per_group_range)
        
        transactions = []
        for j in range(num_transactions_per_group):
            time_delta = pd.Timedelta(minutes=np.random.uniform(1, 10))
            new_date = start_date + time_delta
            
            if j == num_transactions_per_group - 1:
                transaction_status = 2
            else:
                transaction_status = 1
                
            transactions.append({
                'Pan': pan,
                'TransactionType': 3,
                'CreateDate': new_date,
                'TransactionStatus': transaction_status,
                'Amount': start_amount,  # Sabit miktar
                'TransactionRequestDate': new_date + pd.Timedelta(seconds=1),
                'TransactionResponseDate': new_date + pd.Timedelta(seconds=2),
                'CardBrandName': card_brand,
                'CardBankName': card_bank,
                'CardFamily': card_family,
                'CardType': card_type,
                'Fraud': 1  # Tüm grup sahtecilik olarak işaretlenir
            })
        fraud_list.extend(transactions)
    return pd.DataFrame(fraud_list)

# Fraud verileri oluşturma
num_fraud_groups = 3000
num_transactions_per_group_range = (3, 8)
fraud_data = create_fraudulent_groups(df_augmented, num_fraud_groups, num_transactions_per_group_range)

# Tüm verileri birleştirme
df_final = pd.concat([df_augmented, fraud_data], ignore_index=True)
df_final
```
