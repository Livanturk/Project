```python
import pandas as pd
df = pd.read_excel('dataset/veriseti.xlsx')
df.head()
```

```python
#TransactionStatus sütununda eğer 2 varsa onu 1 yapar(başarılı işlem), diğerlerini 0 yapar (başarısız veya beklemedeki işlem)
df['TransactionStatus'] = df['TransactionStatus'].apply(lambda x: 1 if x == 2 else 0)
```

```python
#Tarih içeren sütunları datetime formatına çeviriyoruz
df['CreateDate'] = pd.to_datetime(df['CreateDate'], errors = 'coerce')
df['TransactionRequestDate'] = pd.to_datetime(df['TransactionRequestDate'], errors = 'coerce')
df['TransactionResponseDate'] = pd.to_datetime(df['TransactionResponseDate'], errors = 'coerce')
```

```python
#Tarih içeren sütunların null olan satırlarına forward fill uyguluyoruz
df['TransactionRequestDate'] = df['TransactionRequestDate'].ffill()
df['TransactionResponseDate'] = df['TransactionResponseDate'].ffill()
```

```python
#Date içeriğini elde ediyoruz
df['CreateYear'] = df['CreateDate'].dt.year
df['CreateMonth'] = df['CreateDate'].dt.month
df['CreateDay'] = df['CreateDate'].dt.day
df['CreateHour'] = df['CreateDate'].dt.hour
df['RequestYear'] = df['TransactionRequestDate'].dt.year
df['RequestMonth'] = df['TransactionRequestDate'].dt.month
df['RequestDay'] = df['TransactionRequestDate'].dt.day
df['RequestHour'] = df['TransactionRequestDate'].dt.hour
df['ResponseYear'] = df['TransactionResponseDate'].dt.year
df['ResponseMonth'] = df['TransactionResponseDate'].dt.month
df['ResponseDay'] = df['TransactionResponseDate'].dt.day
df['ResponseHour'] = df['TransactionResponseDate'].dt.hour

#Veri setinde date içeren sütunları atıyoruz
df.drop(['CreateDate', 'TransactionRequestDate', 'TransactionResponseDate'], axis=1, inplace=True)
```

```python
#Kategorik veri içeren sütunlara One-Hot Encoding işlemi uyguluyoruz
df = pd.get_dummies(df, columns = ['CardBrandName', 'CardBankName', 'CardFamily', 'CardType'], drop_first = True)
```
